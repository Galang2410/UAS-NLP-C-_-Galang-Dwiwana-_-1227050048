## Import library ## 

```python
from transformers import pipeline
```

## Load summarization model ##

```python
summarizer = pipeline(
    "summarization",
    model="Irvan14/t5-small-indonesian-summarization",
    tokenizer="Irvan14/t5-small-indonesian-summarization",
    device=-1
)
```

## Baca dataset superclean ##

```python
# membaca dataset
with open("dataset_ai_superclean.txt", "r", encoding="utf-8") as f:
    text = f.read()

```

## Potong dataset menjadi chunk kecil ##

```python
def chunk_text(text, max_length=800):
    words = text.split()
    chunks = []
    current_chunk = []

    for word in words:
        current_chunk.append(word)
        if len(current_chunk) >= max_length:
            chunks.append(" ".join(current_chunk))
            current_chunk = []

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

chunks = chunk_text(text, max_length=800)
len(chunks)
```

## summarization per chunk ##

```python
summaries = []

for i, chunk in enumerate(chunks):
    print(f"Processing chunk {i+1}/{len(chunks)}...")

    try:
        result = summarizer(
            chunk,
            max_length=150,
            min_length=50,
            do_sample=False
        )[0]['summary_text']

        summaries.append(result)

    except Exception as e:
        print("Error chunk", i, ":", e)
        summaries.append("")  # kosongkan jika error
```

```python
final_summary = "\n\n".join(summaries)

with open("summary_output.txt", "w", encoding="utf-8") as f:
    f.write(final_summary)

print("Summary saved to summary_output.txt")
```
